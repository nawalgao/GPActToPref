# Visualize and save all figs for all occupants

```python
syn_occ_data_config_file = '../data/config_files/syn/diff_utilities_config/syn_occupants.json'

with open(syn_occ_data_config_file, 'r') as fd:
    syn_occ_data_config = json.loads(fd.read())

for i in xrange(21):
    occ_num = i + 1
    a = syn_occ_data_config['O' + str(occ_num)]['a']
    b = syn_occ_data_config['O' + str(occ_num)]['b']
    
    # Load actions data
    actions_data = np.loadtxt('../data/syn_data/syn_occ/occupant_duels/train/' +
                                'O_a' + str(a) + '_b' + str(b) + '.csv', delimiter = ',')
    
    actions_data_norm = (actions_data - 0.5)/0.5
    
    X_current = actions_data[:,1]
    X_previous = actions_data[:,0]
    
    loaded_data = np.load('../data/syn_data/tmp_data/predicted_results/' + 'O_a' + str(a) + '_b' + str(b) + '.npz')
    
    grid_levels = loaded_data['Xt1']
    p_mean_grid = loaded_data['p_mean_grid']
    p_var_grid = loaded_data['p_var_grid']
    num_grid = loaded_data['num_grid']
    Xt = loaded_data['Xt']
    u_grid = loaded_data['u_grid']
    
    visualize.contour(grid_levels, X_current, X_previous,
                 p_mean_grid, p_var_grid, a, b, num_grid)
    
    visualize.utility_1D(Xt, u_grid, a, b)
    visualize.utility_1D_samples(Xt, u_grid, a, b)
```
