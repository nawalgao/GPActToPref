```python
for i in xrange(21):
    occ_num = i + 1
    with open(config_file, 'r') as fd:
        config = json.loads(fd.read())

    a = config['O' + str(occ_num)]['a']
    b = config['O' + str(occ_num)]['b']
    
    Occupant = syn_data_gen.OccupantDuels(a,b)
    
    Occupant.save_actions(10, 'train')
    Occupant.save_actions(10, 'test')
    
    Occupant.save_duels(10, 'train')
    Occupant.save_duels(10, 'test')
    
    Occupant.save_utility_plot()
    
    Occupant.save_pref_prob()
    
```
